```python
# Convertir MSRP a número
df['MSRP_num'] = df['MSRP'].replace('[\$,]', '', regex=True).astype(float)

# Agrupar por HorsePower y calcular MSRP promedio
msrp_hp = df.groupby('HorsePower')['MSRP_num'].mean().reset_index()

# Graficar
plt.figure(figsize=(8, 4))
plt.plot(msrp_hp['HorsePower'], msrp_hp['MSRP_num'], marker='o')
plt.title('Relación entre HorsePower y MSRP promedio')
plt.xlabel('HorsePower')
plt.ylabel('MSRP promedio (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

se observa una tendencia creciente entre HorsePower y el MSRP promedio, a medida que aumenta la potencia del motor, el precio sube y a potencias bajas y medias, el aumento es gradual pero despues de los 350 el precio se eleva de manera brusca. indica que los autos mas potentes tienden a ser de lujo o alto rendimiento por lo que el precio es mayor

### Responde en esta celda y a través de código en las siguientes
- Ahora que ya conoces a profundida el dataset realiza IMPUTACIÓN de datos tal como en la tarea 1.5

```python

df_imputado = df.copy()

# Imputar variables numéricas con la mediana
for col in df_imputado.select_dtypes(include=['number']).columns:
    df_imputado[col] = df_imputado[col].fillna(df_imputado[col].median())

# Imputar variables categóricas con la moda
for col in df_imputado.select_dtypes(include='object').columns:
    df_imputado[col] = df_imputado[col].fillna(df_imputado[col].mode()[0])

# Mostrar resumen de valores nulos después de la imputación
print("Valores nulos después de la imputación:")
print(df_imputado.isnull().sum())

```

### Responde en esta celda y a través de código en las siguientes
- Elimina outliers si consideras que afectan al estudio

```python
# Eliminación de outliers usando el método del rango intercuartílico (IQR) para las variables numéricas
df_sin_outliers = df_imputado.copy()
for col in df_numerico.columns:
    Q1 = df_sin_outliers[col].quantile(0.25)
    Q3 = df_sin_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_sin_outliers = df_sin_outliers[(df_sin_outliers[col] >= lower) & (df_sin_outliers[col] <= upper)]

print(f"Filas originales: {df_imputado.shape[0]}")
print(f"Filas después de eliminar outliers: {df_sin_outliers.shape[0]}")
```
